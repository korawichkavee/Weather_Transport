```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import os
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
#load dataset
name = 'Pdataset_20_V04.csv'
df = pd.read_csv(os.getcwd()+"\\Output\\"+name)
df = df[df['DIRECTION']=='Outbound']
df.head()
```

```python
model_name = "model71_OUTOUT" #<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<< MAKE SURE TO CHANGE THE NAME !
#My independent variables
X = df[['lag1_DELAY','lag2_DELAY','lag3_DELAY','lag4_DELAY','WEEKDAY','SCHOOL','T','R_fa','H','W','P']]
Y = df['DELAY']

#fill na with 0
X=X.fillna(0)

#constant at the end
#X = sm.add_constant(X)

model = sm.OLS(Y,X).fit()
#predictions = model.predict(X)

# Print out the statistics
model.summary()
```

```python
X.describe()
```

```python
with open(os.getcwd()+"\\Output\\Saved model\\"+model_name+'summary.txt', 'w') as fh:
    fh.write(model.summary().as_text())
with open(os.getcwd()+"\\Output\\Saved model\\"+model_name+'summary.csv', 'w') as fh:
    fh.write(model.summary().as_csv())
#Save
model.save(os.getcwd()+"\\Output\\Saved model\\"+model_name+"_results.pickle")

print(model.params)
print(model.rsquared)
print(model.rsquared_adj)
print(model.aic)
print(model.bic)
```

```python
#Predict
name = 'Pdataset_10_V04.csv'
df = pd.read_csv(os.getcwd()+"\\Output\\"+name)
df = df[df['DIRECTION']=='Outbound']

#filter 0 delay in y_test
df = df[df['DELAY']!=0]
X_test=df[['lag1_DELAY','lag2_DELAY','lag3_DELAY','lag4_DELAY','WEEKDAY','SCHOOL','T','R_fa','H','W','P']]
#X_test = sm.add_constant(X_test)
y_test=df['DELAY']
y_predict = model.predict(X_test)

#Evaluae
mae = np.mean([abs(true-predict) for true,predict in zip(y_test, y_predict)])
rmse = sqrt(mean_squared_error(y_test, y_predict))
mape = np.mean([abs(true-predict)/true for true,predict in zip(y_test, y_predict)])*100

print('Mean absolute error (MAE): \t\t%.4f min' % mae)
print('Root mean square error (RMSE): \t\t%.4f min' % rmse)
print('Mean absolute percentage error (MAPE): \t%.4f %%' % mape)
```
